```python
U21,I1=genfromtxt("VAC-zond5.csv",delimiter=';',skip_header=1).T
U22,I2=genfromtxt("VAC-zond3.csv",delimiter=';',skip_header=1).T
U23,I3=genfromtxt("VAC-zond1.5.csv",delimiter=';',skip_header=1).T
Ir2=array([5,3,1.5])
len(I3)
```

```python
sU2=.01
sI=.1
errorbar(U21,I1,xerr=sU2,yerr=sI,fmt='.',label=r"$I_p=5$ ")
errorbar(U22,I2,xerr=sU2,yerr=sI,fmt='.',label=r"$I_p=3$ ")
errorbar(U23,I3,xerr=sU2,yerr=sI,fmt='.',label=r"$I_p=1{,}5$ ")
```

```python
from scipy.optimize import curve_fit
def my_curve_fit(f,x,y,sy):
    ko,sko=curve_fit(f,x,y,sigma=sy,absolute_sigma=False)
    sko+=curve_fit(f,x,y,sigma=sy,absolute_sigma=True)[1]
    n=len(sko)
    ssko=zeros(n)
    for i in arange(n):
        ssko[i]=sqrt(sko[i,i])
    for i in arange(n):
        for j in arange(n):
            sko[i,j]=sko[i,j]/(ssko[i]*ssko[j])
    chi2=sum((f(x,ko[0],ko[1],ko[2],ko[3],ko[4])-y)**2/sy**2)
    return ko,ssko,sko,chi2
```

```python
def fitf(x,A,B,C,D,E):
    return A*x+B*tanh(C*(x-D))+E
ko1,sko1,cov1,chi21=my_curve_fit(fitf,U21,I1,sI)
print(ko1,sko1,cov1,chi21,sep='\n')
ko2,sko2,cov2,chi22=my_curve_fit(fitf,U22,I2,sI)
print(ko2,sko2,cov2,chi22,sep='\n')
ko3,sko3,cov3,chi23=my_curve_fit(fitf,U23,I3,sI)
print(ko3,sko3,cov3,chi23,sep='\n')
errorbar(U21,I1,xerr=sU2,yerr=sI,fmt='.',label=r"$I_p=5$ ")
errorbar(U22,I2,xerr=sU2,yerr=sI,fmt='.',label=r"$I_p=3$ ")
errorbar(U23,I3,xerr=sU2,yerr=sI,fmt='.',label=r"$I_p=1{,}5$ ")
arg=arange(-25.5,25.5,.1)
plot(arg,fitf(arg,ko1[0],ko1[1],ko1[2],ko1[3],ko1[4]))
plot(arg,fitf(arg,ko2[0],ko2[1],ko2[2],ko2[3],ko2[4]))
plot(arg,fitf(arg,ko3[0],ko3[1],ko3[2],ko3[3],ko3[4]))
#axis((0,1.2,0,1.2))
minorticks_on()
grid(which='both')
title(" ")
ylabel(r"$I$, ")
xlabel(r"$U_2$, ")
legend(loc='lower right')
savefig("graph1.pdf")
```

$$C=\frac{e}{2kT_e}$$

```python
C=array([ko1[2],ko2[2],ko3[2]])
sC=array([sko1[2],sko2[2],sko3[2]])
Iin=array([ko1[1],ko2[1],ko3[1]])
sIin=array([sko1[1],sko2[1],sko3[1]])
Te=1/2/C
sTe=Te*sC/C
covTC=array([cov1[1,2],cov2[1,2],cov3[1,2]])
print(Te,sTe)
```

$$I_{i} = 0{,}4 n_e eS \sqrt{\frac{2kT_e}{m_i}}$$
$$n_e= \frac{I_{i}}{0{,}4 eS} \sqrt{\frac{m_i}{2kT_e}}$$

```python
ee=1.6e-19
d=.2e-3
sd=.01e-3
l=5.2e-3
sl=.1e-3
mi=22*1.66e-27
ne=Iin*1e-6/(.4*ee*pi*l*d)*sqrt(mi/(2*ee*Te))
sne=ne*sqrt((sl/l)**2+(sd/d)**2+((sIin/Iin)**2+(sTe/Te/2)**2+2*(sIin/Iin)*(sTe/Te/2)*covTC))
print(ne,sne)
```

```python
def olsim(x,y,sx,sy):
	a=0
	for i in range(3):
		w=1/(sy**2+(a*sx)**2)
		Sw=sum(w)
		ax=sum(x*w)/Sw
		ay=sum(y*w)/Sw
		axy=sum(x*y*w)/Sw
		ax2=sum((x**2+sx**2)*w)/Sw
		ay2=sum((y**2+sy**2)*w)/Sw
		a=(axy-ax*ay)/(ax2-ax*ax)
	b=(ax2*ay-ax*axy)/(ax2-ax*ax)
	n=len(x)
	chi2=sum((y-a*x-b)**2/(sy**2+(a*sx)**2))
	s02=n/(n-2)*(ay2-ay*ay-(axy-ax*ay)*(axy-ax*ay)/(ax2-ax*ax))
	sa2=s02/n/(ax2-ax*ax)
	sb2=sa2*ax2
	sa=sqrt(sa2)
	sb=sqrt(sb2)
	return a,b,sa,sb,chi2
```

```python
rc('figure', figsize=(8.27,11.7))
subplot(211)
errorbar(Ir2,Te,xerr=sIr,yerr=sTe,fmt='.')
#axis((0,1.2,0,1.2))
minorticks_on()
grid(which='both')
title(r"     ")
xlabel(r"$I_p$, ")
ylabel(r"$T_e$, ")
#legend(loc='lower right')

an,bn,san,sbn,chi2n=olsim(Ir2,ne,sIr,sne)
print(an,bn,san,sbn,chi2n)
arg2=arange(0,5.5,0.5)

subplot(212)
errorbar(Ir2,ne*1e-16,xerr=sIr,yerr=sne*1e-16,fmt='.')
plot(arg2,(arg2*an+bn)*1e-16)
axis((0,5.5,0,7))
minorticks_on()
grid(which='both')
title(r"     ")
xlabel(r"$I_p$, ")
ylabel(r"$n$, $10^{10}$ $^{-3}$")
#legend(loc='lower right')
tight_layout()
savefig("graph5.pdf")
```
