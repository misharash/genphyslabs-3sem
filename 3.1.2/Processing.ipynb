```python
from scipy import genfromtxt
m,U1,U2,U3=genfromtxt("exp.csv",delimiter=',',skip_header=1).T
len(m)
```

```python
cU=20
sm=10
gamma=3
g=9.8
Umax=1000
d=.16e-2
sd=.01e-2
S=42e-4
sS=1e-4
sU0=sqrt(cU**2+(gamma*Umax*1e-2)**2)
F=g*m*1e-6
sF=g*sm*1e-6
sU0
```

```python
U=(U1+U2+U3)/3
sU=sqrt(sU0**2+((U1-U)**2+(U2-U)**2+(U3-U)**2)/6)
U2=U**2
sU2=2*U*sU
sU
```

```python
def lin0(x,y):
    return x*y
from scipy.optimize import curve_fit
rcParams['figure.figsize']="8,5"
errorbar(F*1e3,U2*1e-4,xerr=sF*1e3,yerr=sU2*1e-4,fmt='.')
ko,sko=curve_fit(lin0,F,U2,sigma=sU2,absolute_sigma=True)
k1,sk1=ko[0],sqrt(sko[0,0])
F1=array(list(F)+[0.0])
plot(F1*1e3,F1*k1*1e-4)
minorticks_on()
grid(which="both")
ylabel("$U^2$, $10^4 \cdot В^2$")
xlabel("$F$, $мН$")
savefig("graph0.pdf",type="pdf",dpi=300)
print("{0:0.2e} {1:0.2e}".format(k1,sk1))
```

```python
eps0=2*d**2/S/k1
seps0=eps0*sqrt((2*sd/d)**2+(sS/S)**2+(sk1/k1)**2)
print("{0:0.2e} {1:0.2e}".format(eps0,seps0))
```

```python
rcParams['figure.figsize']="8,5"
U_CGS=2*d*1e2*sqrt(2*pi*F*1e5/(S*1e4))
sU_CGS=U_CGS*sqrt((sd/d)**2+(sF/F/2)**2+(sS/S/2)**2)
errorbar(U,U_CGS,xerr=sU,yerr=sU_CGS,fmt='.')
ko,sko=curve_fit(lin0,U,U_CGS,sigma=sqrt(sU_CGS**2+(sU/257)**2),absolute_sigma=True)
k2,sk2=ko[0],sqrt(sko[0,0])
U1=array(list(U)+[0.0])
plot(U1,U1*k2)
minorticks_on()
grid(which="both")
ylabel("$U$, $ед.СГС$")
xlabel("$U$, $В$")
axis((0,820,0,3.4))
savefig("graph1.pdf",type="pdf",dpi=300)
print("{0:0.2e} {1:0.2e}".format(k2,sk2))
print("{0:0.2e} {1:0.2e}".format(1/k2,sk2/k2**2))
```
