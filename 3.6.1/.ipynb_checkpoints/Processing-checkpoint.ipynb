```python
import matplotlib.pyplot as plt
from numpy import *
from scipy import genfromtxt
%matplotlib inline
```

```python
def olsim(x,y,sx,sy):
    a=0
    for i in range(3):
        w=1/(sy**2+(a*sx)**2)
        Sw=sum(w)
        ax=sum(x*w)/Sw
        ay=sum(y*w)/Sw
        axy=sum(x*y*w)/Sw
        ax2=sum((x**2+sx**2)*w)/Sw
        ay2=sum((y**2+sy**2)*w)/Sw
        a=(axy-ax*ay)/(ax2-ax*ax)
    b=(ax2*ay-ax*axy)/(ax2-ax*ax)
    n=len(x)
    chi2=sum((y-a*x-b)**2/(sy**2+(a*sx)**2))
    s02=n/(n-2)*(ay2-ay*ay-(axy-ax*ay)*(axy-ax*ay)/(ax2-ax*ax))
    sa2=s02/n/(ax2-ax*ax)
    sb2=sa2*ax2
    sa=sqrt(sa2)
    sb=sqrt(sb2)
    return a,b,sa,sb,chi2
```

```python
tau,dnu=genfromtxt("exp-A.csv",delimiter=";",skip_header=1).T
len(tau)
```

```python
snu=2.5
plt.errorbar(1/tau,dnu,yerr=snu,fmt='.')
a1,b1,sa1,sb1,chi21=olsim(1/tau,dnu,array([0]*8),snu)
plt.plot(1/tau,a1/tau+b1)
print(a1,b1,sa1,sb1,chi21)
```

```python
f,dltnu,N=genfromtxt("exp-B.csv",delimiter=";",skip_header=1).T
len(f)
```

```python
dlnu=dltnu/N
sdlnu=snu/N
plt.errorbar(f,dlnu,yerr=sdlnu,fmt='.')
a2,b2,sa2,sb2,chi22=olsim(f,dlnu,array([0]*6),sdlnu)
plt.plot(f,a2*f+b2)
print(a2,b2,sa2,sb2,chi22)
```

```python
Amax,Amin,aside,amain=genfromtxt("exp-C.csv",delimiter=";",skip_header=1).T
len(Amax)
```

```python
otn=aside/amain
sa=0.2
sotn=otn*sqrt((sa/aside)**2+(sa/amain))
sA=0.1
m=(Amax-Amin)/(Amax+Amin)
sm=m*sqrt((sA/(Amax-Amin))**2+(sA/(Amax+Amin))**2)
```

```python
plt.errorbar(m,otn,yerr=sotn,xerr=sm,fmt='.')
a3,b3,sa3,sb3,chi23=olsim(m,otn,sm,sotn)
plt.plot(m,a3*m+b3)
print(a3,b3,sa3,sb3,chi23)
```

```python
plt.rcParams['figure.figsize']=(14,10)
plt.rcParams['font.size']=16
plt.errorbar(1e3/tau,dnu,yerr=snu,fmt='.')
plt.plot(1e3/tau,a1/tau+b1)
plt.axis((0,45,0,50))
plt.minorticks_on()
plt.grid(which='both')
plt.xlabel(r"$1/\tau, ms^{-1}$")
plt.ylabel(r"$\Delta \nu$, kHz")
plt.savefig("graph0.svg")
```

```python
plt.errorbar(f,dlnu,yerr=sdlnu,fmt='.')
plt.plot(f,a2*f+b2)
plt.axis((0,9,0,10))
plt.minorticks_on()
plt.grid(which='both')
plt.xlabel(r"$f_{повт}$, kHz")
plt.ylabel(r"$\delta \nu$, kHz")
plt.savefig("graph1.svg")
```
