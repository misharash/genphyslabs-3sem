```python
from numpy import *
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from scipy import genfromtxt
R1,x01,x1=genfromtxt("RC.csv",delimiter=":",skip_header=1).T
R1*=1000
len(R1)
```

```python
sx=.1
r=12.2
C=.5e-6
omega=pi*2e3
phi1=-pi*x1/x01
sphi1=phi1*sqrt((sx/x1)**2+(sx/x01)**2)
tphi1=tan(phi1)
stphi1=sphi1/cos(phi1)**2
arg1=1/(omega*C*(R1+r))
```

```python
plt.errorbar(arg1[1:],tphi1[1:],yerr=stphi1[1:],fmt=".")
plt.plot(arg1[1:],arg1[1:])
chi2=sum(((tphi1-arg1)/stphi1)**2)
print(chi2)
```

```python
R2,x02,x2=genfromtxt("RL.csv",delimiter=":",skip_header=1).T
R2*=1000
len(R2)
```

```python
RL=338
L=500e-3
phi2=pi*x2/x02
sphi2=phi2*sqrt((sx/x2)**2+(sx/x02)**2)
tphi2=tan(phi2)
stphi2=sphi2/cos(phi2)**2
arg2=omega*L/(R2+r+RL)
```

```python
plt.errorbar(arg2,tphi2,yerr=stphi2,fmt=".")
plt.plot(arg2,arg2)
chi2=sum(((tphi2-arg2)/stphi2)**2)
print(chi2)
```

```python
f3,x03,x3=genfromtxt("RCL0.csv",delimiter=":",skip_header=1).T
len(f3)
```

```python
C=.05e-6
sf=10
f30=f3[0]
phi3=pi*x3/x03
sphi3=sqrt((sx/x03)**2+(sx*x3/x03**2)**2)
arg3=f3/f30
sarg3=sf/f30
Q3=sqrt(L/C)/(r+RL)
print(Q3)
```

```python
plt.errorbar(arg3,phi3,xerr=sarg3,yerr=sphi3,fmt=".")
plt.plot(arg3,pi/4+0*arg3)
plt.plot(arg3,-pi/4+0*arg3)
```

```python
f4,x04,x4=genfromtxt("RCL100.csv",delimiter=":",skip_header=1).T
len(f4)
```

```python
R=100
f40=f4[0]
phi4=pi*x4/x04
sphi4=sqrt((sx/x04)**2+(sx*x4/x04**2)**2)
arg4=f4/f40
sarg4=sf/f40
Q4=sqrt(L/C)/(r+RL+R)
print(Q4)
```

```python
plt.errorbar(arg4,phi4,xerr=sarg4,yerr=sphi4,fmt=".")
plt.plot(arg4,pi/4+0*arg4)
plt.plot(arg4,-pi/4+0*arg4)
```
